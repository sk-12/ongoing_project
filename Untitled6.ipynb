```
duration=list(data.Duration)
```

```
dur_hrs=[]
dur_mins=[]
for i in range(len(duration)):
  x=duration[i].find('h')
  y=duration[i].find('m')
  if x!=-1:
    dur_hrs.append(int(duration[i][:x]))
  else:
    dur_hrs.append(0)
  if y!=-1:
    if x!=-1:
        dur_mins.append(int(duration[i][x+1:y]))
    else:
        dur_mins.append(int(duration[i][:y]))
  else:
    dur_mins.append(0)
```

```
total_duration=[dur_mins[k]+(dur_hrs[k]*60) for k in range(len(dur_mins))]
```
